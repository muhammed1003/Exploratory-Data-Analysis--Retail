```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load the dataset
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
# check if there is any missing value in any column
df.isnull().sum()
```

```python
# check the data types of the columns
df.dtypes
```

```python
# drop duplicate values
df.drop_duplicates()
```

```python
# Get the count of the unique values of a column
df.nunique()
```

```python
# drop postal code column as it is of no use
df= df.drop ('Postal Code', axis=1)
df
```

```python
# get the column names from the dataframe
df.columns
```

```python
# get the correlation btw variables
df.corr()
```

```python
# get the covariance btw the variables
df.cov()
```

```python
# get the count of unique values from the region columns 
df['Region'].value_counts()
```

```python
sns.set_style ('whitegrid')
sns.countplot(df['Region'],
order=df['Region'].value_counts().index)
```

```python
#get the count of unique values from the state columns
df['State'].value_counts()
```

```python
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.set_style('whitegrid')
sns.countplot(df['State'],
order=df['State'].value_counts().index)
```

```python
# get the count of unique values from the city column
city=df['City'].value_counts()
city.head(10)
```

```python
sns.set_style('whitegrid')
sns.countplot(x=df['Category'],
order=df['Category'].value_counts().index)
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.set_style('whitegrid')
sns.countplot(df['Sub-Category'],
order=df['Sub-Category'].value_counts().index)
```

```python
sns.set_style('whitegrid')
sns.countplot(df['Segment'])
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.set_style('whitegrid') 
sns.barplot(x=df['Sub-Category'],y=df['Sales'])
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.set_style('whitegrid') 
sns.barplot(x=df['State'],y=df['Profit'])
```

```python
top10cities=df.groupby('City')['Sales'].sum().sort_values(ascending=False).iloc[0:10]
plt.title('Sales of Top 10 cities')
plt.xlabel('city')
plt.ylabel('Total sum of sales')
top10cities.plot(kind='bar')
```

```python
top10states=df.groupby('State')['Quantity'].sum().sort_values(ascending=False).iloc[0:10]
plt.title('Sales of Top 10 states')
plt.xlabel('city')
plt.ylabel('Total sum of sales')
top10states.plot(kind='bar')
```

```python
# count of discount sales grouped by city
discount=df.groupby('City')['Discount'].value_counts().sort_values(ascending=False).iloc[0:10]
discount
```

```python
# statewise sales and profit overview
overview=df.groupby('State')['Profit','Sales'].sum()
overview.plot(kind='bar',figsize=(25,8))
```
